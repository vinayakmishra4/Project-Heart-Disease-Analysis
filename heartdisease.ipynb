```python
for column in df.columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
# Univariate Analysis
continuous_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

for col in continuous_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
sns.boxplot(x='target', y='age', data=df)
plt.title("Age vs Target")
plt.show()
```

```python
sns.boxplot(x='target', y='chol', data=df)
plt.title("Cholesterol vs Target")
plt.show()
```

```python
sns.boxplot(x='target', y='thalach', data=df)
plt.title("Max Heart Rate vs Target")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

Strongest positive predictors of heart disease are chest pain type (cp) and max heart rate (thalach).

Exercise-induced angina (exang), oldpeak (ST depression), and number of vessels (ca) negatively correlate with heart disease.

Age and sex show weak negative correlations with the target.

Most other features have weak correlations, indicating they might be less important individually.
